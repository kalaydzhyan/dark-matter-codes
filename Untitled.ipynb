```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, num=1000, endpoint=True)
y = np.cos(-x**2/9.0)
f = interpolate.interp1d(x, y)
y1 = f(x)
plt.plot(x, y1)
```

```python
fp = np.gradient(y)
plt.plot(x, fp)
```
